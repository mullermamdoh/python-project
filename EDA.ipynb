```python
import pandas as pd
import ast

# read the cleaned csv 
df_movies = pd.read_csv("movies_cleaned.csv")
```

### 1. Top 10 highest-rated movies

```python
top_10 = df_movies.sort_values(by='vote_average', ascending=False).head(10)
print(top_10[['title', 'vote_average']])
```

```python
# with vote_count condition
# مع شرط ان ال vote_count > 500 

top_10 = df_movies[df_movies['vote_count'] > 500] \
            .sort_values(by='vote_average', ascending=False) \
            .head(10)

print(top_10[['title', 'vote_average', 'vote_count']])
```

## 2. Count of movies per genre

```python
df_movies['genres'].value_counts()
```

```python
# convert genres from str to list so explode can work
df_movies['genres'] = df_movies['genres'].apply(ast.literal_eval)

print(type(df_movies['genres'].iloc[0]))
```

### 8. Count of movies per genre

```python
df_genre = df_movies.explode('genres')

genre_count = df_genre['genres'].value_counts()

print(genre_count)
```

### 3. Filter movies released before 2000

```python
movies_before_2000 = df_movies[df_movies['release_date'] < 2000]

print(movies_before_2000[['title', 'release_date']].head())
print("Total movies before 2000:", len(movies_before_2000))
```

### 4. Movies with rating above average

```python
avg_rating = df_movies['vote_average'].mean()

above_avg_movies = df_movies[df_movies['vote_average'] > avg_rating]

print("Average rating:", avg_rating)
print(above_avg_movies[['title', 'vote_average']].head())
print("Total movies above average:", len(above_avg_movies))
```
