### 13) Evolution of Film Runtime Over the Years (10 Films Sampled per Year)





```python
def sample_n_per_year(df, n=10):
    return df.groupby("release_date").apply(lambda x: x.sample(n=min(len(x), n), random_state=42)).reset_index(drop=True)

sampled_df = sample_n_per_year(df, n=10)

avg_runtime_sampled = sampled_df.groupby("release_date")["runtime"].mean().sort_index()

plt.figure(figsize=(12,6))
plt.plot(avg_runtime_sampled.index, avg_runtime_sampled.values, marker='o', color='blue')
plt.title("Evolution of Film Runtime Over the Years (10 Films Sampled per Year)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Runtime (minutes)", fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Only keep films from the last 100 years
current_year = pd.Timestamp.now().year
df_recent = df[df['release_date'] >= current_year - 100]

# Function to sample up to n films per year
def sample_n_per_year(df, n=10):
    return df.groupby("release_date").apply(lambda x: x.sample(n=min(len(x), n), random_state=42)).reset_index(drop=True)

# Sample 10 films per year
sampled_df = sample_n_per_year(df_recent, n=10)

# Compute average runtime per year from sampled data
avg_runtime_sampled = sampled_df.groupby("release_date")["runtime"].mean().sort_index()

# Optional: 5-year rolling average for smoothing
avg_runtime_rolling = avg_runtime_sampled.rolling(window=5, min_periods=1).mean()

# Plot
plt.figure(figsize=(14,6))
plt.plot(avg_runtime_sampled.index, avg_runtime_sampled.values, 'o', alpha=0.5, label='Yearly Sample Mean')
plt.plot(avg_runtime_rolling.index, avg_runtime_rolling.values, '-', color='blue', label='5-Year Rolling Avg')
plt.title("Evolution of Film Runtime Over the Last 100 Years (10 Films Sampled per Year)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Runtime (minutes)", fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
