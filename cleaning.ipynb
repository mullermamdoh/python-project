## Movies_metadata.csv dataset Cleaning and exploration

```python
import pandas as pd
import ast
```

```python
# read csv files
df_movies = pd.read_csv(r"D:\ITI Power BI\Python\rating\movies_metadata.csv")
```

```python
# data exploration 

print (df_movies.info())
```

```python
# compute null rows
print(df_movies.isnull().sum())
print(df_movies['overview'])
print(df_movies['runtime'])
```

```python
### Data cleaning
# drop unnecessary colums
columns_to_drop = [
    'belongs_to_collection',
    'homepage',
    'imdb_id',
    'poster_path',
    'production_companies',
    'production_countries',
    'spoken_languages',
    'status',
    'video',
    'adult',
    'tagline',
    'original_title',
    'revenue',
    'budget'
]
df_movies = df_movies.drop(columns=columns_to_drop, errors='ignore')
print(df_movies['popularity'])
```

```python
# convert to numeric values
df_movies['popularity'] = pd.to_numeric(df_movies['popularity'], errors='coerce')
```

```python
# remove all rows with null values
df_movies = df_movies.dropna(subset=['title', 'vote_average', 'vote_count', 'genres'])
```

```python
# Fill numeric columns with median
df_movies['runtime'] = df_movies['runtime'].fillna(df_movies['runtime'].median())# runtime=duration
df_movies['popularity'] = df_movies['popularity'].fillna(df_movies['popularity'].median())
```

```python
# Fill text fields
df_movies['overview'] = df_movies['overview'].fillna("No overview")#overview = review
```

```python
# Fill categorical fields
df_movies['original_language'] = df_movies['original_language'].fillna("unknown")
```

```python
# Handle release_date
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], errors='coerce')# release date= year
df_movies['release_date'] = df_movies['release_date'].dt.year.fillna(0).astype(int)
```

```python
# export genres name only
def parse_genres(x):
    try:
        genres_list = ast.literal_eval(x)
        return [g['name'] for g in genres_list]
    
    except:
        return []
        
df_movies['genres'] = df_movies['genres'].apply(parse_genres)
```

```python
# Replace empty lists with 'Unknown'
df_movies['genres'] = df_movies['genres'].apply(lambda x: x if x else ['Unknown'])
print (df_movies.info())
print(df_movies.columns)
```
